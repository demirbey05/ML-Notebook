# Comparison between Deep Neural Networks and Shallow Neural Networks

For shallow network with scalar input x and scalar output y with D hidden units, the # linear region is equal to D+1 and number of parameters is equal to 3D+1. In Deep NN, with same input and output and K layer, # of linear region is equal to (D+1) ^ K and number of parameters is 3D + 1 + D (D+1)(K-1)


Now we will compare the ratio of linear region to total parameters for different D values. D>2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
D_values = np.arange(2,10)
K_values = [2,3,4,5,6]
```

```python
def calculate_ratios_dl(D_values, K_values):
    ratios = {K: [] for K in K_values}
    
    for K in K_values:
        for D in D_values:
        # Deep network
            deep_linear_regions = (D + 1) ** K
            deep_parameters = 3 * D + 1 + D * (D + 1) * (K - 1)
            deep_ratio = deep_linear_regions / deep_parameters
            ratios[K].append(deep_ratio)

    return ratios


def calculate_ratios_sh(D_values):
    ratios = []
    
    for D in D_values:
        shallow_regions = (D + 1)
        shallow_parameters = 3 * D + 1
        deep_ratio = shallow_regions / shallow_parameters
        ratios[K].append(shallow_ratio)

    return ratios
```

```python
ratios = calculate_ratios_dl(D_values, K_values)

# Plot the results
plt.figure(figsize=(12, 8))
colors = plt.cm.viridis(np.linspace(0, 1, len(K_values)))

for i, K in enumerate(K_values):
    plt.plot(D_values, ratios[K], label=f'K = {K}', color=colors[i], marker='o')

plt.xlabel('D (Number of Hidden Units)')
plt.ylabel('Ratio of Linear Regions to Total Parameters')
plt.title('Comparison of Ratios for Different K Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
ratios = calculate_ratios_sh(D_values)

# Plot the results
plt.figure(figsize=(12, 8))
plt.plot(D_values, ratios, marker='o')

plt.xlabel('D (Number of Hidden Units)')
plt.ylabel('Ratio of Linear Regions to Total Parameters')
plt.title('Comparison of Ratios for Different K Values')
plt.grid(True)
plt.show()
```
