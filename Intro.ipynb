```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Â Create dataset using sine function
num_points = 1000
x_sine = np.linspace(-20 * np.pi, 20 * np.pi, num_points)
noise = np.random.normal(0, 0.5, num_points)  # Gaussian noise

y_sine = np.sin(x_sine) + noise
```

```python
## Create dataset using linear function
num_points = 100
x_linear = np.linspace(-20 * np.pi, 20 * np.pi, num_points)
slope = 0.5
intercept = 2
noise_linear = np.random.normal(0, 2, num_points)  # Gaussian noise

y_linear = slope * x_linear + intercept + noise_linear
```

```python
# Plotting
fig, axs = plt.subplots(2, 1, figsize=(12, 10))

# Sine plot
axs[0].scatter(x_sine, y_sine, label='Data with Noise', color='b', s=10, alpha=0.6)
axs[0].plot(x_sine, np.sin(x_sine), label='True Function: sin(x)', color='r', linestyle='--')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('Synthetic Data with Sine Function')
axs[0].legend()
axs[0].grid(True)

# Linear plot
axs[1].scatter(x_linear, y_linear, label='Linear Data with Noise', color='g', s=10, alpha=0.6)
axs[1].plot(x_linear, slope * x_linear + intercept, label='True Line: y = 0.5x + 2', color='orange', linestyle='--')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_title('Synthetic Linear Data for 1D Linear Regression')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
class LinearRegression:
    def __init__(self):
        self.theta_zero = np.random.rand(1)
        self.theta_one = np.random.rand(1)

    def compute(self,x:float) ->float:
        return x*self.theta_one + self.theta_zero

    def find_optimal_parameters(self,x:np.ndarray,y:np.ndarray):
        n = x.size
        base = (n*np.sum(x*x)) - (np.sum(x)**2)

        self.theta_zero = ((np.sum(x*x)*np.sum(y)) - (np.sum(x*y)*np.sum(x)))/ base
        self.theta_one = ((n*np.sum(x*y)) - (np.sum(y)*np.sum(x)))/base
```

```python
regressor = LinearRegression()
```

```python
regressor.find_optimal_parameters(x_linear,y_linear)
```

```python
print(regressor.theta_zero) # true value : 2
print(regressor.theta_one) # true value: 0.5
```

```python
class GenLinearRegression:
    def __init__(self):
        self.theta_zero = np.random.rand(1)
        self.theta_one = np.random.rand(1)

    def compute(self,x:float) ->float:
        return (x- self.theta_zero)/self.theta_one

    def find_optimal_parameters(self,x:np.ndarray,y:np.ndarray):
        n = x.size
        base = (n*np.sum(y*y)) - (np.sum(y)**2)

        self.theta_zero = ((np.sum(y*y)*np.sum(x)) - (np.sum(x*y)*np.sum(y)))/ base
        self.theta_one = ((n*np.sum(x*y)) - (np.sum(x)*np.sum(y)))/base
```

```python
regressor = GenLinearRegression()
```

```python
regressor.find_optimal_parameters(x_linear,y_linear)
```

```python
print(regressor.theta_zero) # true value : 2
print(regressor.theta_one) # true value: 0.5
```
